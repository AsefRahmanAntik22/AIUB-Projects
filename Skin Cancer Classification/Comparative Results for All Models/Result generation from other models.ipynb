```python
# Import required libraries
import matplotlib.pyplot as plt  # For plotting
import numpy as np               # For numerical operations

# Define the metrics and phases (used throughout the radar plots)
metrics = ["accuracy", "precision", "recall", "f1_score"]  # Common performance metrics
phases = ["train", "test"]  # Data phases
best_model_names = [] # for storing best test accuracy model from every category of the model
```

```python
# Define a function to generate radar charts for filtered models and selected data phase
def plot_radar_filtered(models_data, 
                        phase,                     # Phase of the dataset: 'train', 'validation', or 'test'
                        filter_keyword=None,       # Keyword(s) to filter model names (can be a string or list of strings)
                        set_ylim=[0.6, 1.0],       # Y-axis limits for radar chart
                        set_yticks=[0.6, 0.7, 0.8, 0.9, 1.0],  # Y-axis tick marks
                        fig_size=(5, 5)):          # Size of the radar chart figure

    # Set the metric labels for radar chart axes
    labels = metrics
    num_vars = len(labels)  # Number of axes = number of metrics

    # Compute angle for each axis in the radar chart
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += [angles[0]]  # Repeat the first angle at the end to complete the circular plot

    # Create a polar plot for the radar chart
    fig, ax = plt.subplots(figsize=fig_size, subplot_kw=dict(polar=True))

    # === Filter models by keywords ===
    if filter_keyword:
        # If the user passed a single keyword (string), convert it to a list for uniform processing
        if isinstance(filter_keyword, str):
            filter_keyword = [filter_keyword]
        
        # Filter the models whose names contain ANY of the keywords (case-insensitive match)
        filtered_models = {
            k: v for k, v in models_data.items()
            if any(keyword.lower() in k.lower() for keyword in filter_keyword)
        }
    else:
        # If no filter keywords provided, include all models
        filtered_models = models_data

    # If no models match the filter, print a warning and return None
    if not filtered_models:
        print(f"No models found with keywords: {filter_keyword}")
        return None

    # Variables to keep track of the model with the highest accuracy
    best_model = None
    best_accuracy = -1

    # === Plot radar for each filtered model ===
    for model_name, model_info in filtered_models.items():
        # Extract values for each metric in the selected phase (e.g., 'test')
        values = [model_info[phase][metric] for metric in labels]
        values += [values[0]]  # Repeat the first value to close the polygon

        # Draw the polygon line for the model
        ax.plot(angles, values, label=model_name)

        # Optionally fill the radar area under the curve with transparency
        ax.fill(angles, values, alpha=0.05)

        # Update best model if this one has a higher accuracy
        current_accuracy = model_info[phase]["accuracy"]
        if current_accuracy > best_accuracy:
            best_accuracy = current_accuracy
            best_model = model_name

    # === Radar chart formatting and aesthetics ===
    
    # Set the chart title
    ax.set_title(f"{phase.capitalize()} Metrics Radar Chart - Filter: {filter_keyword}", size=12, y=1.08)

    # Set the labels (performance metrics) around the radar axes
    ax.set_xticks(angles[:-1])              # Exclude the duplicate angle used to close the loop
    ax.set_xticklabels(labels)             # Metric names as axis labels

    # Set the radial axis (performance metric values)
    ax.set_yticks(set_yticks)               # Radial ticks (grid lines)
    ax.set_ylim(set_ylim[0], set_ylim[1])  # Y-axis range

    # Enable grid for readability
    ax.grid(True)

    # Add legend to identify each model, outside the chart
    ax.legend(loc='center left', bbox_to_anchor=(0.9, 0.9), fontsize=8)

    # Automatically adjust layout to avoid clipping
    plt.tight_layout()

    # Show the radar chart
    plt.show()

    # Return the name of the model with the highest accuracy
    return best_model
```

```python
filter_keyword = "conv"
set_ylim=[0.6, 1.0]
set_yticks=[0.6, 0.7, 0.8, 0.9, 1.0]
fig_size = (5,5)


for phase in phases:
    if phase=='test':
        best_model_names.append(plot_radar_filtered(models_data, phase, filter_keyword, set_ylim=set_ylim, set_yticks=set_yticks, fig_size = fig_size))
    else:
        plot_radar_filtered(models_data, phase, filter_keyword, set_ylim=set_ylim, set_yticks=set_yticks, fig_size = fig_size)
```
