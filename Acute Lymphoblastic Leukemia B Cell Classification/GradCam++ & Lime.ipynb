```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Reshape, Multiply
from tensorflow.keras.models import Model
from lime import lime_image
from skimage.segmentation import mark_boundaries
import pickle

with open("Merged Dataset/stacking_model.pkl", "rb") as file:
    stacking = pickle.load(file)

input_shape = (224, 224, 3)
base_model = VGG16(weights="imagenet", include_top=False, input_shape=input_shape)
gap = GlobalAveragePooling2D()(base_model.output)
se = Dense(base_model.output_shape[-1] // 16, activation="relu")(gap)
se = Dense(base_model.output_shape[-1], activation="sigmoid")(se)
se = Reshape((1, 1, -1))(se)
x = Multiply()([base_model.output, se])
x = GlobalAveragePooling2D()(x)  
feature_extractor = Model(inputs=base_model.input, outputs=x)

image_path = "C:/Users/Faysal Ahmmed/Desktop/ALL classification/Merged Dataset/Datasets/normalized_output of Blood Cancer vahadane/Benign/aug_5808.jpg"
#image_path = "C:/Users/Faysal Ahmmed/Desktop/ALL classification/Dataset 2/Datasets/normalized_output of Blood Cancer vahadane/Benign/aug_504.jpg" 
original_image = cv2.imread(image_path)
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
resized_image = cv2.resize(original_image, (224, 224))
input_image = preprocess_input(np.expand_dims(resized_image, axis=0))

def gradcam_plus_plus(model, img_array, layer_name="block5_conv3"):
    grad_model = tf.keras.models.Model(
        [model.inputs], [model.get_layer(layer_name).output, model.output]
    )

    with tf.GradientTape() as tape:
        conv_outputs, predictions = grad_model(img_array)
        loss = tf.reduce_max(predictions)  # Get max prediction class

    grads = tape.gradient(loss, conv_outputs)  # Compute gradients
    grads = tf.nn.relu(grads)  # Ensure positive gradients
    weights = tf.reduce_mean(grads, axis=(0, 1, 2))  # Global average pooling
    grad_cam = np.sum(weights * conv_outputs.numpy()[0], axis=-1)

    grad_cam = np.maximum(grad_cam, 0)  # ReLU
    grad_cam = cv2.resize(grad_cam, (224, 224))  # Resize to original size
    grad_cam = grad_cam / np.max(grad_cam)  # Normalize

    heatmap = cv2.applyColorMap(np.uint8(255 * grad_cam), cv2.COLORMAP_JET)
    heatmap = cv2.cvtColor(heatmap, cv2.COLOR_BGR2RGB)
    superimposed_img = cv2.addWeighted(resized_image, 0.6, heatmap, 0.4, 0)

    return superimposed_img

gradcam_result = gradcam_plus_plus(base_model, input_image)

explainer = lime_image.LimeImageExplainer()

def predict_fn(images):
    images = preprocess_input(images)
    extracted_features = feature_extractor.predict(images) 
    return stacking.predict_proba(extracted_features) 

explanation = explainer.explain_instance(
    resized_image.astype(np.double),
    predict_fn,
    top_labels=2,
    hide_color=0,
    num_samples=300
)

lime_result, mask = explanation.get_image_and_mask(
    explanation.top_labels[0], positive_only=True, num_features=10, hide_rest=True
)
lime_result = mark_boundaries(lime_result, mask)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(original_image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(gradcam_result)
ax[1].set_title("Grad-CAM++")
ax[1].axis("off")

ax[2].imshow(lime_result)
ax[2].set_title("LIME")
ax[2].axis("off")

plt.show()
```
